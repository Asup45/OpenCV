## Operations Simples

```python
# use opencv to load and display the image
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage.transform as transform
import imageio.v2 as imageio

# lire l'image (avec imread)
image = imageio.imread("image_01.png")

# afficher l'image
plt.imshow(image)
plt.title("Image originale")
plt.show()
```

```python
# afficher la taille de l'image
print(f"Taille de l'image: {image.shape}")  # (hauteur, largeur, canaux)
```

```python
# l'image a trois couleur R,G,B. afficher les valeurs des trois couleurs pour le pixel x = 150, et y = 100
x, y = 150, 100
print(f"Valeurs RGB pour le pixel (x=150, y=100): {image[y, x]}")
```

```python
# Crop une partie de l'image (exemple: coin supérieur gauche)
cropped = image[50:150, 50:150]  # (y1:y2, x1:x2)
plt.imshow(cropped)
plt.title("Image croppée")
plt.show()
```

```python
# redimensioner l'image à 200x200 pixels
resized = cv2.resize(image, (200, 200)) 
plt.imshow(resized)
plt.title("Image redimensionnée à 200x200")
plt.show()
```

```python
# faire tourner une image -45°
rotated = transform.rotate(image, -45, resize=True, mode='edge')
plt.imshow(rotated)
plt.title("Image tournée de -45°")
plt.show()
```

# Filtrage d'image

* Blur
* Gaussian blur
* Median blur
* Sharpening
* Bilateral blur
* Bilateral filtering

```python
# charger et afficher l'image "bois"
bois = imageio.imread("bois.png")
plt.imshow(bois)
```

```python
# Utiliser la méthode "cv2.filter2D" avec un kernel de taille 2
kernel_2x2 = np.ones((2, 2), np.float32) / 4
filter2D = cv2.filter2D(bois, -1, kernel_2x2)
plt.subplot(232)
plt.imshow(filter2D)
plt.title('Custom Filter 2x2')
```

```python
# Appliquer blur avec un kernel de taille 5x5
blur = cv2.blur(bois, (5, 5))
plt.subplot(233)
plt.imshow(blur)
plt.title('Blur 5x5')
```

```python
# gaussian blur
gaussian_blur = cv2.GaussianBlur(bois, (5, 5), 0)
plt.subplot(234)
plt.imshow(gaussian_blur)
plt.title('Gaussian Blur')
```

```python
# median blur
median_blur = cv2.medianBlur(bois, 5)
plt.subplot(235)
plt.imshow(median_blur)
plt.title('Median Blur')
```

```python
# sharpening
sharpening_kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
sharpened = cv2.filter2D(bois, -1, sharpening_kernel)
plt.subplot(236)
plt.imshow(sharpened)
plt.title('Sharpening')

plt.tight_layout()
plt.show()
```

```python
# bilateral filtering
bilateral = cv2.bilateralFilter(bois, 9, 75, 75)
plt.figure(figsize=(6, 4))
plt.imshow(bilateral)
plt.title('Bilateral Filtering')
plt.show()
```

# Seuillage d'image

En utilisant l'image "sudoku" appliquer:
* Binary Thresholding
* Otsu thresholding
* Adaptive thresholding

```python
# charger et afficher l'image "sudoku"
sudoku = imageio.imread("sudoku.jpg")
# Convertir l'image en niveau de gris
gray_sudoku = cv2.cvtColor(sudoku, cv2.COLOR_RGB2GRAY)
# On redimensionne l'image en 200x200
gray_sudoku_resized = cv2.resize(gray_sudoku, (200, 200)) 
plt.imshow(gray_sudoku_resized)
```

```python
# appliquer un seuillage binaire d'un seuil de 127
_, thresh_127 = cv2.threshold(gray_sudoku_resized, 127, 255, cv2.THRESH_BINARY)
```

```python
# appliquer un seuillage binaire d'un seuil de 30
_, thresh_30 = cv2.threshold(gray_sudoku_resized, 30, 255, cv2.THRESH_BINARY)
```

```python
# appliquer un seuillage binaire d'un seuil de 240
_, thresh_240 = cv2.threshold(gray_sudoku_resized, 240, 255, cv2.THRESH_BINARY)
```

Qu'est ce que vous remarquez ?!

```python
# appliquer "Otsu" seuillage
_, otsu_thresh = cv2.threshold(gray_sudoku_resized, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
# appliquer "Adaptive threshold"
adaptive_thresh = cv2.adaptiveThreshold(gray_sudoku_resized, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
```

```python
# Afficher les résultats
plt.figure(figsize=(10, 8))
plt.subplot(321), plt.imshow(sudoku), plt.title('Original')
plt.subplot(322), plt.imshow(thresh_127, cmap='gray'), plt.title('Seuillage 127')
plt.subplot(323), plt.imshow(thresh_30, cmap='gray'), plt.title('Seuillage 30')
plt.subplot(324), plt.imshow(thresh_240, cmap='gray'), plt.title('Seuillage 240')
plt.subplot(325), plt.imshow(otsu_thresh, cmap='gray'), plt.title('Otsu Seuillage')
plt.subplot(326), plt.imshow(adaptive_thresh, cmap='gray'), plt.title('Seuillage Adaptatif')
plt.tight_layout()
plt.show()
```

# Détection des bords

* Sobel
* Canny

```python
# charger et afficher l'image "image_02"
image_02 = imageio.imread("image_02.png")
# Convertir l'image en niveau de gris
gray_image_02 = cv2.cvtColor(image_02, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image_02)
```

```python
# utiliser la methode sobel pour détecter les bords
# Vous pouvez utiliser d'autres images de votre choix
sobelx = cv2.Sobel(gray_image_02, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(gray_image_02, cv2.CV_64F, 0, 1, ksize=3)
sobel_edges = cv2.magnitude(sobelx, sobely)
```

```python
# Canny
canny_edges = cv2.Canny(gray_image_02, 100, 200)
```

# Détection des contours

```python
# faites une recherche sur internet pour détecter les contours des objets dans image_02
contours, _ = cv2.findContours(canny_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

#  Compter des objets

```python
# Compter les objets
number_of_objects = len(contours)
```

```python
# Afficher les résultats
plt.figure(figsize=(12, 10))
plt.subplot(221), plt.imshow(sobel_edges, cmap='gray'), plt.title('Sobel Edges')
plt.subplot(222), plt.imshow(canny_edges, cmap='gray'), plt.title('Canny Edges')
plt.subplot(223), plt.imshow(cv2.drawContours(gray_image_02.copy(), contours, -1, (0, 255, 0), 3)), plt.title(f'Contours (Objects Found: {number_of_objects})')
plt.tight_layout()
plt.show()
```
